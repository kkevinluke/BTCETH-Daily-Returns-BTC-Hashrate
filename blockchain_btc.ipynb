```python
# The size of each step in days
day_delta = datetime.timedelta(days=1)

start_date = datetime.date.today()
end_date = start_date - 30*day_delta
dates=[]
for i in range((start_date - end_date).days):
    x=start_date - i*day_delta
    dates.append(x.strftime("%Y-%m-%d"))
```

```python
df_new["dates"] = dates
```

```python
df_new.drop("time", axis=1)
```

```python
fig, ax = plt.subplots(figsize=(28,7))
ax2 = ax.twinx()
ax.set_title('Bitcoin Hashrate and Price')
ax.set_xlabel('Days')
ax.plot(df_new['dates'], df_new['hashrate'], color='green', marker='x')
ax2.plot(df_new['dates'], df_new['open_price'], color='red', marker='o')
#ax.set_ylabel('GDP Per Capita (US $)')
#ax2.set_ylabel('Annual Growth Rate (%)')
ax.legend(['Hashrate'])
ax2.legend(['Open Price'], loc='upper center')
#ax.set_xticks(gdp['date'].dt.date)
#ax.set_xticklabels(gdp['date'].dt.year, rotation=90)
ax.yaxis.grid(color='lightgray', linestyle='dashed')
plt.tight_layout()
plt.show()
```
