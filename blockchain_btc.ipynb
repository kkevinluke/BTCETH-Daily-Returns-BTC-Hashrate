```python
df_new["open_price"] = df_btc["open"]

df_new["high_price"] = df_btc["high"]
```

```python
df_new.head()
```

```python
len(df_new)
```

```python
# The size of each step in days
day_delta = datetime.timedelta(days=1)

start_date = datetime.date.today()
end_date = start_date - 30*day_delta
dates=[]
for i in range((start_date - end_date).days):
    x=start_date - i*day_delta
    dates.append(x.strftime("%Y-%m-%d"))
```

```python
df_new["dates"] = dates
```

```python
df_new["hashrate"].describe()
```

```python
fig = plt.figure(figsize=(28,7))
ax = fig.add_subplot(111)
ax2 = ax.twinx()

l1 = ax.plot(df_new['dates'], df_new['hashrate'], color='green', marker='x', label="Hashrate")
l2 =ax2.plot(df_new['dates'], df_new['open_price'], color='red', marker='o', label="Open Price")

lines = l1 + l2
labels = [l.get_label() for l in lines]
ax.legend(lines, labels, loc=1)

ax.set_title('Bitcoin Hashrate and Open Price')
ax.set_xlabel('Days')
ax.set_ylabel("Hashrate")
ax2.set_ylabel('Open Price')
ax.grid(color='lightgray', linestyle='dashed')
plt.tight_layout()
plt.show()
```
